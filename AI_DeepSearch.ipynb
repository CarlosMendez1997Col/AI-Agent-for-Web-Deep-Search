# Ministerio de Ciencia Tecnología e Innovación

```python
def corregir_minciencias():
    base_url = "https://minciencias.gov.co"
    url_objetivo = "https://minciencias.gov.co/convocatorias/todas"
    
    print(f"{'='*20} MINCIENCIAS: {'='*20}")
    
    try:
        resp = requests.get(url_objetivo, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        
        # Lista temporal para esta institucion
        datos_minciencias = []
        
        # Buscamos en la tabla de convocatorias
        filas = soup.find_all("tr")
        for fila in filas:
            cols = fila.find_all("td")
            if len(cols) >= 2:
                link_tag = cols[1].find("a")
                if link_tag and link_tag.has_attr('href'):
                    # LIMPIEZA DE URL: urljoin combina la base con el link relativo
                    url_final = urljoin(base_url, link_tag['href'])
                    titulo = link_tag.get_text(strip=True)
                    
                    # Verificación visual
                    print(f"Institucion: MinCiencias")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    
                    datos_minciencias.append({
                        "institucion": "MinCiencias",
                        "titulo_convocatoria": titulo,
                        "url_externa": url_final
                    })
        
        return datos_minciencias

    except Exception as e:
        print(f"Error al corregir MinCiencias: {e}")
        return []

# Ejecutamos la primera
data_1 = corregir_minciencias()
```

# Ministerio de Ambiente y Desarrollo Sostenible

```python
def corregir_minambiente():
    # URL de la sección de convocatorias
    url_objetivo = "https://www.minambiente.gov.co/convocatorias"
    # El dominio base para completar links relativos
    base_url = "https://www.minambiente.gov.co"
    
    print(f"{'='*20} MINAMBIENTE: {'='*20}")
    
    try:
        resp = requests.get(url_objetivo, timeout=10)
        # Usamos lxml o html.parser para procesar el contenido
        soup = BeautifulSoup(resp.text, "html.parser")
        
        datos_minambiente = []
        
        # MinAmbiente suele organizar esto en tablas. 
        # Buscamos todas las filas que tengan enlaces.
        enlaces = soup.select("table td a")
        
        if not enlaces:
            print("No se encontraron enlaces en la tabla. Verificando estructura alternativa...")
            # Búsqueda genérica por si la estructura cambió
            enlaces = soup.find_all("a", href=True)

        for link in enlaces:
            href = link.get('href')
            titulo = link.get_text(strip=True)
            
            # Filtramos para evitar capturar el menú de navegación o links vacíos
            if href and len(titulo) > 10: 
                # Reconstrucción de la URL
                url_final = urljoin(base_url, href)
                
                print(f"Institucion: MinAmbiente")
                print(f"Titulo: {titulo}")
                print(f"Url: {url_final}")
                print("-" * 50)
                
                datos_minambiente.append({
                    "institucion": "MinAmbiente",
                    "titulo_convocatoria": titulo,
                    "url_externa": url_final
                })
        
        return datos_minambiente

    except Exception as e:
        print(f"Error al corregir MinAmbiente: {e}")
        return []

# Ejecutamos la segunda
data_2 = corregir_minambiente()
```

# Fundación Natura Colombia

```python
def corregir_natura_final():
    url_objetivo = "https://natura.org.co/convocatorias/"
    base_url = "https://natura.org.co"
    
    print(f"{'='*20} FUNDACIÓN NATURA: {'='*20}")
    
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        
        datos_natura = []
        enlaces_vistos = set()

        for a in soup.find_all("a", href=True):
            href = a['href']
            
            # Solo procesamos si es una página de convocatoria y no la hemos visto
            if "convocatoria" in href.lower() and href not in enlaces_vistos:
                url_final = urljoin(base_url, href)
                
                # --- PROCESO DE DECODIFICACIÓN EXCLUSIVO ---
                # Extraemos la parte final de la URL (ej: convocatoria-profesional-calidad)
                slug = urlparse(url_final).path.strip('/')
                
                # Limpieza: quitamos guiones, quitamos la palabra "convocatoria" y formateamos
                titulo_limpio = slug.replace('-', ' ').lower()
                titulo_limpio = titulo_limpio.replace('convocatoria', '').strip().capitalize()
                
                # Verificación visual
                print(f"Institucion: Fundación Natura")
                print(f"Titulo (Decodificado): {titulo_limpio}")
                print(f"Url: {url_final}")
                print("-" * 50)
                
                datos_natura.append({
                    "institucion": "Fundación Natura",
                    "titulo_convocatoria": titulo_limpio,
                    "url_externa": url_final
                })
                enlaces_vistos.add(href)

        return datos_natura

    except Exception as e:
        print(f"Error al corregir Fundación Natura: {e}")
        return []

# Ejecutamos la validación
data_3 = corregir_natura_final()
```

# Fundación para la Conservación y el Desarrollo

```python
def corregir_fcds_v2():
    url_objetivo = "https://fcds.org.co/oportunidades-laborales/"
    base_url = "https://fcds.org.co"
    
    print(f"{'='*20} FCDS: {'='*20}")
    
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        
        datos_fcds = []
        enlaces_vistos = set()

        # Buscamos todos los enlaces que contengan la frase que me indicaste
        # o que apunten a la carpeta de uploads donde están los PDFs
        enlaces = soup.find_all("a", href=True)

        for a in enlaces:
            href = a['href']
            texto_boton = a.get_text(strip=True)
            
            # Criterio: El enlace debe ser un PDF y estar en uploads, 
            # o el texto debe decir "Más información"
            if (".pdf" in href.lower() and "uploads" in href.lower()) or ("más información" in texto_boton.lower()):
                
                if href not in enlaces_vistos:
                    url_final = urljoin(base_url, href)
                    
                    # --- DECODIFICACIÓN DEL TÍTULO DESDE EL NOMBRE DEL ARCHIVO ---
                    # Extraemos el nombre del archivo (ej: tdr-coordinadorproyectobid-fcds-2026.pdf)
                    nombre_archivo = urlparse(url_final).path.split('/')[-1]
                    
                    # Limpiamos el nombre: quitamos .pdf, guiones y palabras técnicas
                    titulo_decodificado = nombre_archivo.replace('.pdf', '').replace('-', ' ').replace('_', ' ')
                    titulo_decodificado = titulo_decodificado.replace('tdr', 'TDR:').strip().capitalize()
                    
                    print(f"Institucion: FCDS")
                    print(f"Titulo: {titulo_decodificado}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    
                    datos_fcds.append({
                        "institucion": "FCDS",
                        "titulo_convocatoria": titulo_decodificado,
                        "url_externa": url_final
                    })
                    enlaces_vistos.add(href)
        
        return datos_fcds

    except Exception as e:
        print(f"Error al corregir FCDS: {e}")
        return []

# Ejecutamos la validación
data_4 = corregir_fcds_v2()
```

# FontAgro

```python
def corregir_fontagro():
    # URL específica de las convocatorias
    url_objetivo = "https://fontagro.org/es/iniciativas/convocatorias"
    # Dominio base para completar enlaces si es necesario
    base_url = "https://fontagro.org"
    
    print(f"{'='*20} FONTAGRO: {'='*20}")
    
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        
        datos_fontagro = []
        
        # FontAgro suele usar la clase 'initiative-title' para los encabezados h4
        tarjetas = soup.find_all("h4", class_="initiative-title")
        
        for titulo_tag in tarjetas:
            # El título está en el texto del h4
            titulo = titulo_tag.get_text(strip=True)
            
            # Buscamos el enlace 'a' que suele estar dentro del h4 o inmediatamente después
            link_tag = titulo_tag.find("a") if titulo_tag.find("a") else titulo_tag.find_next("a")
            
            if link_tag and link_tag.has_attr('href'):
                href = link_tag['href']
                # Construimos la URL absoluta
                url_final = urljoin(base_url, href)
                
                print(f"Institucion: FontAgro")
                print(f"Titulo: {titulo}")
                print(f"Url: {url_final}")
                print("-" * 50)
                
                datos_fontagro.append({
                    "institucion": "FontAgro",
                    "titulo_convocatoria": titulo,
                    "url_externa": url_final
                })
        
        if not datos_fontagro:
            print("Aviso: No se detectaron convocatorias con los selectores actuales. Verificando estructura...")

        return datos_fontagro

    except Exception as e:
        print(f"Error al corregir FontAgro: {e}")
        return []

# Ejecutamos la validación para la quinta institución
data_5 = corregir_fontagro()
```

# NESsT

```python
def corregir_nesst():
    url_objetivo = "https://nesst.bamboohr.com/careers"
    
    print(f"{'='*20} NESsT: {'='*20}")
    
    try:
        # 1. Navegamos a la URL
        driver.get(url_objetivo)
        
        # 2. Esperamos hasta que aparezcan los enlaces de las vacantes
        # BambooHR suele usar la clase 'fab-LinkUnstyled' para los títulos de vacantes
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "a.fab-LinkUnstyled"))
        )
        
        # 3. Pasamos el contenido renderizado a BeautifulSoup
        soup = BeautifulSoup(driver.page_source, "html.parser")
        
        datos_nesst = []
        enlaces = soup.select("a.fab-LinkUnstyled")
        
        for link in enlaces:
            titulo = link.get_text(strip=True)
            href = link.get('href')
            
            if href:
                # BambooHR suele dar URLs completas, pero por seguridad las validamos
                url_final = href if href.startswith('http') else f"https://nesst.bamboohr.com{href}"
                
                print(f"Institucion: NESsT")
                print(f"Titulo: {titulo}")
                print(f"Url: {url_final}")
                print("-" * 50)
                
                datos_nesst.append({
                    "institucion": "NESsT",
                    "titulo_convocatoria": titulo,
                    "url_externa": url_final
                })
        
        if not datos_nesst:
            print("Aviso: No se encontraron vacantes activas en este momento.")
            
        return datos_nesst

    except Exception as e:
        print(f"Error al corregir NESsT: {e}")
        return []

# Ejecutamos la validación para la sexta institución
# Nota: Asegúrate de tener el objeto 'driver' activo de los pasos anteriores
data_6 = corregir_nesst()
```

# Programa Colombia Sostenible

```python
def corregir_colombia_sostenible():
    url_objetivo = "https://www.colombiasostenible.gov.co/programa/convocatorias/"
    base_url = "https://www.colombiasostenible.gov.co"
    
    print(f"{'='*20} COLOMBIA SOSTENIBLE: {'='*20}")
    
    try:
        driver.get(url_objetivo)
        
        # Esperamos a que la estructura principal de las convocatorias cargue
        # Suelen usar una clase llamada 'rt-detail' o simplemente títulos dentro de la sección de entrada
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "entry-title"))
        )
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_colombia = []
        
        # Buscamos los bloques de convocatoria
        articulos = soup.find_all("h3", class_="entry-title")
        
        if not articulos:
            # Si no los encuentra por h3, buscamos cualquier enlace dentro del contenido principal
            articulos = soup.select(".entry-content a")

        for item in articulos:
            # Si el item es un h3, buscamos el 'a' adentro; si es un 'a', lo usamos directamente
            link_tag = item if item.name == "a" else item.find("a")
            
            if link_tag and link_tag.has_attr('href'):
                titulo = link_tag.get_text(strip=True)
                href = link_tag['href']
                
                # Filtramos ruidos cortos o enlaces de navegación
                if len(titulo) > 10:
                    url_final = urljoin(base_url, href)
                    
                    print(f"Institucion: Colombia Sostenible")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    
                    datos_colombia.append({
                        "institucion": "Colombia Sostenible",
                        "titulo_convocatoria": titulo,
                        "url_externa": url_final
                    })
        
        return datos_colombia

    except Exception as e:
        print(f"Error al corregir Colombia Sostenible: {e}")
        return []

# Ejecutamos la validación para la séptima
data_7 = corregir_colombia_sostenible()
```

# Fondo Acción

```python
def corregir_fondo_accion_v2():
    url_objetivo = "https://fondoaccion.org/convocatorias/"
    
    print(f"{'='*20} FONDO ACCIÓN: {'='*20}")
    
    try:
        driver.get(url_objetivo)
        
        # Esperamos a que la galería cargue
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "fg-item-title"))
        )
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_fondo = []
        enlaces_vistos = set()

        # Apuntamos directamente a los contenedores de los títulos
        titulos_bloques = soup.find_all("h2", class_="fg-item-title")

        for h2 in titulos_bloques:
            # El título real es el texto del h2
            titulo = h2.get_text(strip=True)
            
            # Buscamos el enlace 'a' que envuelve al h2 o que es su padre directo
            # Según tu imagen, el h2 está DENTRO de un <a>
            link_tag = h2.find_parent("a")
            
            if link_tag and link_tag.has_attr('href'):
                url_real = link_tag['href']
                
                # Evitamos duplicados y links de imágenes (asegurándonos que no termine en .jpg)
                if url_real not in enlaces_vistos and not url_real.lower().endswith(('.jpg', '.jpeg', '.png')):
                    
                    print(f"Institucion: Fondo Acción")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_real}")
                    print("-" * 50)
                    
                    datos_fondo.append({
                        "institucion": "Fondo Acción",
                        "titulo_convocatoria": titulo,
                        "url_externa": url_real
                    })
                    enlaces_vistos.add(url_real)
        
        return datos_fondo

    except Exception as e:
        print(f"Error al corregir Fondo Acción: {e}")
        return []

# Ejecutamos la validación corregida
data_8 = corregir_fondo_accion_v2()
```

# Patrimonio Natural

```python
def corregir_patrimonio_natural():
    url_objetivo = "https://patrimonionatural.org.co/convocatorias/"
    base_url = "https://patrimonionatural.org.co"
    
    print(f"{'='*20} PATRIMONIO NATURAL: {'='*20}")
    
    try:
        driver.get(url_objetivo)
        
        # Esperamos a que carguen los títulos principales (suelen ser h2 o h3)
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "elementor-heading-title"))
        )
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_patrimonio = []
        enlaces_vistos = set()

        # Buscamos los títulos que suelen tener la clase de Elementor
        titulos = soup.find_all(["h2", "h3"], class_="elementor-heading-title")

        for tag in titulos:
            link_tag = tag.find("a")
            if link_tag and link_tag.has_attr('href'):
                titulo = link_tag.get_text(strip=True)
                url_real = urljoin(base_url, link_tag['href'])
                
                # Filtramos para que el título sea relevante y la URL no sea una imagen
                if len(titulo) > 10 and not url_real.lower().endswith(('.jpg', '.png', '.jpeg')):
                    if url_real not in enlaces_vistos:
                        print(f"Institucion: Patrimonio Natural")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_real}")
                        print("-" * 50)
                        
                        datos_patrimonio.append({
                            "institucion": "Patrimonio Natural",
                            "titulo_convocatoria": titulo,
                            "url_externa": url_real
                        })
                        enlaces_vistos.add(url_real)
        
        return datos_patrimonio

    except Exception as e:
        print(f"Error al corregir Patrimonio Natural: {e}")
        return []

# Ejecutamos la validación para la novena y última
data_9 = corregir_patrimonio_natural()
```

# Banco de Desarrollo de América Latina (CAF)

```python
def extraer_caf_final():
    # Definimos las rutas confirmadas
    urls_objetivo = [
        "https://www.caf.com/es/trabaja-con-nosotros/convocatorias/",
        "https://www.caf.com/es/trabaja-con-nosotros/licitaciones/"
    ]
    base_url = "https://www.caf.com"
    
    print(f"{'='*20} CAF: (Convocatorias + Licitaciones) {'='*20}")
    
    datos_caf = []
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

    for url in urls_objetivo:
        try:
            resp = requests.get(url, headers=headers, timeout=15)
            soup = BeautifulSoup(resp.text, "html.parser")
            
            # Buscamos todos los H3 que contienen títulos según tus capturas
            # Seleccionamos tanto la clase de Convocatorias como la de Licitaciones
            encabezados_titulo = soup.find_all("h3", class_=["card__title", "card--gris-50__title"])
            
            for h3 in encabezados_titulo:
                link_tag = h3.find("a", href=True)
                
                if link_tag:
                    titulo = link_tag.get_text(strip=True)
                    href = link_tag['href']
                    url_final = urljoin(base_url, href)
                    
                    # Filtro de Seguridad y Relevancia (Habeas Data)
                    if len(titulo) > 10 and "mailto:" not in url_final.lower():
                        # Limpieza de duplicados de texto (ej. si el título trae códigos de licitación)
                        print(f"Institucion: CAF")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_final}")
                        print("-" * 50)
                        
                        datos_caf.append({
                            "Institucion": "CAF",
                            "Titulo": titulo,
                            "Url": url_final
                        })
                        
        except Exception as e:
            print(f"Error procesando {url}: {e}")
            
    return datos_caf

# Ejecución para la base de datos de carlosmendez1997col
data_caf_total = extraer_caf_final()
```

```python
import os
import pandas as pd
from bs4 import BeautifulSoup

def extract_full_text(html_path):
    """Extrae el texto plano del HTML eliminando ruido técnico."""
    try:
        with open(html_path, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')
            for element in soup(["script", "style"]):
                element.decompose()
            return soup.get_text(separator=' ', strip=True).replace("'", "''") # Escapar comillas simples para SQL
    except Exception as e:
        print(f"Error procesando {html_path}: {e}")
        return None

# 1. Preparación de los datos
base_folder = 'CONVOCATORIAS_HTML'
output_sql_file = 'SEI_GENA_Postgres_Dump.sql'

print("🚀 Generando archivo SQL para PostgreSQL...")

# 2. Escribir el encabezado del archivo SQL
with open(output_sql_file, 'w', encoding='utf-8') as f:
    f.write("-- SEI-GENA: Base de Datos de Convocatorias\n")
    f.write("DROP TABLE IF EXISTS convocatorias;\n")
    f.write("CREATE TABLE convocatorias (\n")
    f.write("    institucion TEXT,\n")
    f.write("    titulo TEXT,\n")
    f.write("    url TEXT,\n")
    f.write("    informacion TEXT\n")
    f.write(");\n\n")

    # 3. Procesar archivos e insertar registros
    count = 0
    for institucion in os.listdir(base_folder):
        inst_path = os.path.join(base_folder, institucion)
        if os.path.isdir(inst_path):
            print(f"📦 Extrayendo: {institucion}")
            for archivo in os.listdir(inst_path):
                if archivo.endswith(".html"):
                    full_path = os.path.join(inst_path, archivo)
                    
                    titulo = archivo.replace(".html", "").replace("_", " ").replace("'", "''")
                    url_ref = f"Ref: {institucion}/{archivo}"
                    contenido = extract_full_text(full_path)
                    
                    if contenido:
                        # Crear la sentencia INSERT
                        sql_insert = f"INSERT INTO convocatorias (institucion, titulo, url, informacion) VALUES ('{institucion}', '{titulo}', '{url_ref}', '{contenido}');\n"
                        f.write(sql_insert)
                        count += 1

print(f"\n✅ ¡Proceso finalizado! Se ha creado el archivo: {output_sql_file}")
print(f"📊 {count} registros exportados al script SQL.")
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# --- Base de datos ---
df = pd.DataFrame(convocatorias)

# --- Embeddings ---
model = SentenceTransformer("all-MiniLM-L6-v2")
texts = df["titulo_convocatoria"] + " " + df["objetivo_descripcion"].fillna("")
embeddings = model.encode(texts.tolist())

# --- Historial ---
historial = []

# --- Función de respuesta ---
def responder(pregunta, top_k=2):
    historial.append(("usuario", pregunta))

    pregunta_emb = model.encode([pregunta])
    sims = cosine_similarity(pregunta_emb, embeddings)[0]
    indices = sims.argsort()[-top_k:][::-1]
    resultados = df.iloc[indices]

    respuesta = "He encontrado estas convocatorias relacionadas:\n\n"
    for _, row in resultados.iterrows():
        respuesta += f"- {row['institucion']}: {row['titulo_convocatoria']}\n"
        if row['objetivo_descripcion']:
            respuesta += f"  Objetivo: {row['objetivo_descripcion']}\n"
        if row['fecha_inicio'] or row['fecha_cierre']:
            respuesta += f"  Fechas: {row['fecha_inicio']} - {row['fecha_cierre']}\n"
        if row['recursos']:
            respuesta += f"  Recursos: {row['recursos']}\n"
        respuesta += f"  Más info: {row['link']}\n\n"

    historial.append(("asistente", respuesta))
    return respuesta

# --- Widgets estilo Chat ---
entrada = widgets.Text(
    value='',
    placeholder='Escribe tu pregunta aquí...',
    description='Prompt:',
    layout=widgets.Layout(width='80%')
)

salida = widgets.Output()

# --- Botón para limpiar historial ---
boton_limpiar = widgets.Button(
    description="Limpiar historial",
    button_style="danger"
)

def mostrar_historial():
    salida.clear_output()
    with salida:
        for rol, contenido in historial:
            if rol == "usuario":
                display(widgets.HTML(f"<div style='background-color:#DCF8C6; padding:10px; border-radius:10px; margin:5px; text-align:right'><b>Tú:</b> {contenido}</div>"))
            else:
                display(widgets.HTML(f"<div style='background-color:#E6E6E6; padding:10px; border-radius:10px; margin:5px; text-align:left'><b>Asistente:</b> {contenido}</div>"))

def on_submit(change):
    if change.new.strip():
        responder(change.new)
        mostrar_historial()
        entrada.value = ""  # limpiar campo de texto

def limpiar_historial(b):
    historial.clear()
    mostrar_historial()

entrada.observe(on_submit, names="value")
boton_limpiar.on_click(limpiar_historial)

display(entrada, boton_limpiar, salida)

```
